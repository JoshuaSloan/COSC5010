```python
# import all required libraries
import numpy as np
import pandas as pd
import string
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# load the finefoods dataset
#https://stackoverflow.com/questions/49070668/amazon-fine-foods-convert-text-file-to-pandas-dataframe
with open('finefoods.txt','r', encoding='latin-1') as f:
    data = f.read()
columns_list=['product/productId',
 'review/helpfulness',
 'review/profileName',
 'review/score',
 'review/summary',
 'review/text',
 'review/time',
 'review/userId']

df=pd.DataFrame([
    {line.split(': ')[0]:': '.join(line.split(': ')[1:]) 
     for line in record.split('\n') if line.split(': ')[0] in columns_list}
    for record in data.split('\n\n')])
```

```python
df.head()
```

```python
#Remove all HTML tags from the data (failing to do so will allow a large volume of HTML tag words like "br" and "p")
reviewText = df['review/text'].astype(str).str.replace('<[^<]+?>', '').to_numpy()
print(reviewText.shape)
```

```python
# read the long stop word list from the .txt file
with open("long_stop_word_list.txt", encoding="utf16") as f2:
    contents = f2.read()
```

```python
# load the stop words into a list 
stopList = []
for word in contents.split():
    stopList.append(word)
print(len(stopList))
```

```python
# check to ensure that all stop words were loaded properly
for i in range(len(stopList)):
    print(i,stopList[i])
```

```python
# remove all punctutation from the data
remove = string.punctuation
remove = remove.replace("'", "") # don't remove apostrophes
for i in range(len(reviewText)):
    reviewText[i] = reviewText[i].translate(str.maketrans('', '', remove))
```

```python
# make all words lowercase such that comparisons are NOT case-sensitive
words = []
for text in reviewText:
    for w in text.split():
        words.append(w.lower())
```

```python
# print the total number of words in "review/text" after filtering
print(len(words))
```

```python
# print all unique words in "review/text"
L = set(words)
print(len(L))
```

```python
# find all stop-words in unique word list
words2remove = []
for i in L:
    if i in stopList:
        words2remove.append(i)

# print the number of stop words found
print(len(words2remove))

# remove stop words from the list
for j in words2remove:
    L.remove(j)  
# print resulting word list to check that words were properly removed
print(len(L)) 
```

```python
# initializing a dictionary
d = {};

# count the number of times each word comes up in list of words
for key in words: 
    d[key] = d.get(key, 0) + 1

countW = sorted(d.items(), key = lambda x: x[1], reverse = True)
```

```python
# get the top 500 words and their respective count
top500 = []
for i in range(len(countW)):
    if(countW[i][0] in L and len(top500) < 500):
        top500.append(countW[i])
```

```python
# print the topp 500 words (+ their respective counts)
print(top500)
```

```python
# remove counts for the next step
top500Words = []
for i in range(len(top500)):
    top500Words.append(top500[i][0])
```

```python
# create CountVectorizer (using the stop list)
vectorizer = CountVectorizer(stop_words=stopList)
vectorizer.fit(top500Words)

# check that the vocab list matches the top 500 words
print(vectorizer.vocabulary_)

# get top 500 word counts from filtered "review/text"
vector = vectorizer.transform(reviewText)
```

```python
# train a 10-means classifier from the previous result
# (to decrease the run-time, lower the value of max_iter)
kmeans = KMeans(n_clusters=10, random_state=0, init='k-means++', max_iter=300, n_init=1)
kmeans.fit(vector.toarray())
```

```python
# print out the top 5 words for each cluster and their respective counts
print("Top terms per cluster:")
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(10):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :5]:
        print('%s' % terms[ind],"count:",order_centroids[i,ind]),
    print()
```
